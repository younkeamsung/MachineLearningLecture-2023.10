```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')
```

```python
sentences = ['안녕하세요?', '한글 문장 임베딩을 위한 BERT 모델입니다.']
embeddings = model.encode(sentences)
print(embeddings)
```

```python
df = pd.read_csv('data/부동산정책data.csv')
df.head(3)
```

```python
model.encode(df.유저[0])
```

```python
df['embedding'] = pd.Series([[]] * len(df))     # dummy
df.shape
```

```python
df.embedding = df.유저.map(lambda x: list(model.encode(x)))
df.head(3)
```

```python
df.to_csv('data/부동정책_dataset.csv', index=False)
```

```python
import json
wdf = pd.read_csv('data/부동정책_dataset.csv')
wdf.embedding = wdf.embedding.apply(json.loads)
wdf.head(3)
```

```python
text = '전세'
embedding = model.encode(text)
```

```python
wdf['유사도'] = wdf.embedding.map(lambda x: cosine_similarity([embedding],[x]).squeeze())
```

```python
wdf.유사도.idxmax()
```

```python
print(wdf.내용[wdf.유사도.idxmax()])
```

```python
text = '청년'
embedding = model.encode(text)
wdf['유사도'] = wdf.embedding.map(lambda x: cosine_similarity([embedding], [x]).squeeze())
answer = wdf.내용[wdf.유사도.idxmax()]
print(answer)
```
